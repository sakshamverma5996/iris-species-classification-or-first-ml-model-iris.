```python
def p(test_values, result):
    correct_values = 0
    wrong = 0
    
    # zip() pairs up each actual value with its corresponding predicted value
    for actual, predicted in zip(test_values, result):
        if actual == predicted:
            print("true")
            correct_values += 1
        else:
            print("false")
            wrong += 1
            
    print(f"\nTotal Correct: {correct_values}")
    print(f"Total Wrong: {wrong}")
    
    return correct_values, wrong
```
